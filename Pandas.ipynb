# Styling dataframe

## Adding Background gradient

```python
import pandas as pd
import numpy as np
```

```python
arr = np.random.randint(1,10,50).reshape(10,5)
df = pd.DataFrame(data = arr,columns=['A','B','C','D','E'])
```

```python
#Adding same background style to all columns. This applies to only numeric columns
df.style.background_gradient(cmap = 'Greens')
```

```python
#Adding same style to particular columns only.
df.style.background_gradient(cmap = 'Blues', subset = ['A','E'])
```

```python
#Adding different style to different columns
#'\'' is used to continue the line in next row
df.style.background_gradient(cmap='Blues', subset = ['A', 'D'])\
.background_gradient(cmap = 'Greens', subset = ['B', 'E'])\
.background_gradient(cmap = 'Reds', subset = 'C')
```
